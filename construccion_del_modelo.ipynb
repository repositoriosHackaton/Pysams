```python
#transformar datos
from sklearn.preprocessing import StandardScaler
```

```python
datos = pd.read_csv("cardio_train_normalizado.csv", delimiter=",")
df = pd.DataFrame(datos)
```

```python
df.head()
```

```python
variables_numericas = datos[['age','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active']]
```

```python
variables_numericas.head()
```

```python
#nombres de las columnas
nuevas_columnas = {0:'age',1:'gender',2:'height',3:'weight',4:'ap_hi',5:'ap_lo',6:'cholesterol',7:'gluc',8:'smoke',9:'alco',10:'active'}
```

```python
#escalar datos
scaler = StandardScaler()
variables_normalizadas = scaler.fit_transform(variables_numericas)
```

```python
df_n = pd.DataFrame(variables_normalizadas)
df_n.rename(columns=nuevas_columnas, inplace=True)
```

```python
df_n
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
data = df_n[['age','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active']]
target = df['cardio']
```

```python
target
```

```python
#separacion de los datos, en datos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size=0.25, random_state=25)
```

```python
#modelo
modelo = RandomForestClassifier(n_estimators=500, max_depth=10, min_samples_split=10, random_state=25)
```

```python
#entrenamiento
modelo.fit(X_train, Y_train)
```

```python
#prediccion
y_pred = modelo.predict(X_test)
```

```python
#evaluacion de la precision
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test,y_pred)
print('La precision del modelo es ', accuracy)
```

```python
# Persistencia del modelo

import joblib

# guardar el modelo
joblib.dump(modelo, "random_forest.pkl")


# revisar si se guard el archivo
!ls

# descargar ese archivo
from google.colab import files
files.download("random_forest.pkl")
print("Descargado con xito")
```
