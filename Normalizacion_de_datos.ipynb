```python
df = pd.read_csv('cardio_train.csv', sep=';')
# Conversion de la edad en días a años y redondear a enteros
df['age'] = (df['age'] / 365.25).round().astype(int)
```

```python
df.head()
```

```python
#Características y variable objetivo
features = df.drop(columns=['id', 'cardio'])
target = df['cardio']
```

```python
#Escalador
scaler = MinMaxScaler()
```

```python
# Ajuste y transformacion de las características
normalized_features = scaler.fit_transform(features)
```

```python
normalized_data = pd.DataFrame(normalized_features, columns=features.columns)
```

```python
normalized_data['cardio'] = target.values
```
